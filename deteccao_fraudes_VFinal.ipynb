```python
#verifica os os tipos de dados nao fraude
df_nao_fraude = df_credit[df_credit.Class == 0]
df_nao_fraude
```

```python
#dataset com fraudes com a mesma quantidade de nao fraudes
df_nao_fraudes = df_nao_fraude.sample(n=492)
df_nao_fraudes
```

```python
#concatenando os dois dataframes nao fraude e fraude
df = pd.concat([df_nao_fraudes, df_fraude], axis=0)
```

```python
#ajustando o index 
df.reset_index(inplace=True)
df
```

```python
#retirar dados para validacao
df_val_nao_fraude = df.head(5)
df_val_nao_fraude 
```

```python
#capturando os cinco ultimos registros fraude para validacao
df_val_fraude  = df.tail(5)
df_val_fraude  
```

```python
#retirar as cinco primeiras linhas
df  = df.iloc[5:]
```

```python
# retirar as cinco as cinco ultimas linhas 
df  = df[:-5]

df
```

```python
#concatenando o dataframe de validacao
df_val_total = pd.concat([df_val_nao_fraude, df_val_fraude])

df_val_total.reset_index(inplace=True)
df_total_real = df_val_total.Class
df_val_total = df_val_total.drop(['level_0', 'index', 'Time', 'Class'], axis=1)
df_val_total
```

```python
# verificar a distribuicao dos dados do tipo fraude e nao fraude
df.Class.value_counts()
```

```python
#separar Labels e Features
X= df.drop(['index', 'Time', 'Class'], axis=1)
Y = df['Class']

print(X, Y)
```

```python
#separar entre os dados de treino e teste
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state = 42,  stratify=Y)
```

```python
#Treianmento
lr  =LogisticRegression(max_iter=1000)
lr.fit(X_train, Y_train)
pred = lr.predict(X_test)
acc = accuracy_score(Y_test, pred)

f'Acurácia:{acc * 100:2f}'
```

```python
#Validacao
pred = lr.predict(df_val_total)

df = pd.DataFrame({'real':df_total_real, 'previsao':pred})

print(df)
```
